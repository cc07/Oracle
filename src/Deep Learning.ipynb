```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train.drop('Survived', axis=1), y_train, test_size=0.3, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=0)

print('Training set', X_train.shape, y_train.shape)
print('Validation set', X_valid.shape, y_valid.shape)
print('Test set', X_test.shape, y_test.shape)
```

```python
def accuracy(predictions, labels):
    correct_pred = tf.equal(tf.argmax(predictions, 1), tf.argmax(labels, 1))
    return tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

```python
_startLearningRate = 0.0035
_learningDecayRate = 0.985
_decaySteps = X_train.shape[0]
training_epochs = 5000
batch_size = 100
display_step = 1
beta = 0.03

n_feature = X_train.shape[1]
n_output = y_train.shape[1]
n_hidden = 16

train_writer = tf.summary.FileWriter('./tmp/log/train', graph=tf.get_default_graph())
test_writer = tf.summary.FileWriter('./tmp/log/test')

graph = tf.Graph()
with graph.as_default():
    
    is_training = tf.placeholder(tf.bool, shape=())
    dataset = tf.placeholder(tf.float32, shape=(None, n_feature))
    labels = tf.placeholder(tf.float32, shape=(None, n_output))

    keep_prob = 0.5 if is_training == True else 1.    
    
    w1 = tf.get_variable("w1", shape=[n_feature, n_hidden], initializer=tf.contrib.layers.xavier_initializer())
    b1 = tf.get_variable('biases1', [n_hidden], initializer=tf.random_normal_initializer())
    w1_hist = tf.summary.histogram("weights1", w1)
    b1_hist = tf.summary.histogram("biases1", b1)
    
    layer1 = tf.nn.bias_add(tf.matmul(dataset, w1), b1)
#     layer1 = tf.layers.batch_normalization(layer1, training=is_training, axis=1)
    layer1 = tf.nn.relu(layer1)
    dropout1 = tf.nn.dropout(layer1, keep_prob)

    w2 = tf.get_variable("w2", shape=[n_hidden, n_output], initializer=tf.contrib.layers.xavier_initializer())
    b2 = tf.get_variable('biases2', [n_output], initializer=tf.random_normal_initializer())
    w2_hist = tf.summary.histogram("weights2", w2)
    b2_hist = tf.summary.histogram("biases2", b2)
    logits = tf.nn.bias_add(tf.matmul(dropout1, w2), b2)    
#     layer2 = tf.nn.bias_add(tf.matmul(dropout1, w2), b2)    
#     dropout2 = tf.nn.dropout(layer2, keep_prob)
    
#     w3 = tf.get_variable("w3", shape=[n_hidden, n_hidden], initializer=tf.contrib.layers.xavier_initializer())
#     b3 = tf.get_variable('biases3', [n_hidden], initializer=tf.random_normal_initializer())
#     w3_hist = tf.summary.histogram("weights3", w3)
#     b3_hist = tf.summary.histogram("biases3", b3)
#     layer3 = tf.nn.bias_add(tf.matmul(dropout2, w3), b3)        
#     dropout3 = tf.nn.dropout(layer3, keep_prob)
    
#     w4 = tf.get_variable("w4", shape=[n_hidden, n_output], initializer=tf.contrib.layers.xavier_initializer())
#     b4 = tf.get_variable('biases4', [n_output], initializer=tf.random_normal_initializer())
#     w4_hist = tf.summary.histogram("weights4", w4)
#     b4_hist = tf.summary.histogram("biases4", b4)
#     logits = tf.nn.bias_add(tf.matmul(dropout3, w4), b4)        
    
    
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=labels, logits=logits))
    
    prediction = tf.nn.softmax(logits)    
    regularizer = tf.nn.l2_loss(w1) + tf.nn.l2_loss(b1)
    regularizer += tf.nn.l2_loss(w2) + tf.nn.l2_loss(b2)
    regularizer += tf.nn.l2_loss(w3) + tf.nn.l2_loss(b3)
    regularizer += tf.nn.l2_loss(w4) + tf.nn.l2_loss(b4)    
    loss = tf.reduce_mean(loss + beta * regularizer)
    tf.summary.scalar("loss", loss)
    
#     global_step = tf.cast(tf.Variable(0), tf.int64)
#     learning_rate = tf.train.exponential_decay(_startLearningRate, global_step, _decaySteps, _learningDecayRate)
    learning_rate = _startLearningRate
    optimizer = tf.train.AdagradOptimizer(learning_rate=learning_rate).minimize(loss)
#     optimizer = tf.train.ProximalAdagradOptimizer(learning_rate=learning_rate).minimize(loss)
    
    acc = accuracy(prediction, labels)    
    tf.summary.scalar('accuracy', acc)
    
    merged = tf.summary.merge_all()
```

```python
n_stat = []
acc_batch = []
acc_valid = []
loss_batch = []
loss_valid = []

with tf.Session(graph=graph) as sess:
    
    tf.global_variables_initializer().run()
    
    for epoch in range(training_epochs):
        
        avg_cost = 0
        avg_acc = 0
        total_batch = int(X_train.shape[0] / batch_size)
        
        X_batches = np.array_split(X_train.values, total_batch)
        y_batches = np.array_split(y_train, total_batch)
        
        for i in range(total_batch):
            
            X_batch, y_batch = X_batches[i], y_batches[i]
            
            summary, batch_acc_rate, _, cost = sess.run([merged, acc, optimizer, loss], feed_dict={ dataset: X_batch, labels: y_batch, is_training: True })
            train_writer.add_summary(summary, epoch * total_batch + i)
            summary, valid_acc_rate, valid_cost = sess.run([merged, acc, loss], feed_dict={ dataset: X_valid.values, labels: y_valid, is_training: False })            
            test_writer.add_summary(summary, epoch * total_batch + i)
            
#             batch_prediction, _, cost = sess.run([prediction, optimizer, loss], feed_dict={ dataset: X_batch, labels: y_batch })
#             valid_prediction, valid_cost = sess.run([prediction, loss], feed_dict={ dataset: X_valid.values, labels: y_valid, is_training: False })
            
#             batch_acc_rate = accuracy(batch_prediction, y_batch)
#             valid_acc_rate = accuracy(valid_prediction, y_valid)
            avg_cost += cost / total_batch
            avg_acc += batch_acc_rate / total_batch
            
#             print('Accuracy: Batch=%.4f, Valid=%.4f' % 
#                   (batch_acc_rate, valid_acc_rate))                 
            
        if epoch % display_step == 0:
            print('Epoch %04d: cost=%.4f' % (epoch, avg_cost))
            
        n_stat.append(epoch)
        acc_batch.append(avg_acc)
        acc_valid.append(valid_acc_rate)
        loss_batch.append(avg_cost)
        loss_valid.append(valid_cost)                    
        
        test_acc = sess.run(acc, feed_dict={ dataset: X_test.values, labels: y_test, is_training: False })

        print('Accuracy: Train=%.4f, Valid=%.4f, Test=%.4f' % 
              (avg_acc, valid_acc_rate, test_acc))                     

#         test_prediction = sess.run(prediction, feed_dict={ dataset: X_test.values, labels: y_test, is_training: False })

#         print('Accuracy: Train=%.4f, Valid=%.4f, Test=%.4f' % 
#               (avg_acc, valid_acc_rate, accuracy(test_prediction, y_test)))             

        avg = np.mean([avg_acc, valid_acc_rate, test_acc])
        if (avg > best):
            print('Above previous best, saving...')
            best = avg
            saver = tf.train.Saver()
            saver.save(sess, "./data/model.ckpt")        
```
